### Selecting rows from DataFrame

* iloc: searches using index position (zero indexing)
* loc: searches using index labels. (custom indexing) (both start and end are inclusive)

Note: incase of index of df is default than both iloc[1:3], and loc[1:3] but only differnence is in loc, both are inclusive unlike iloc.
i.e loc will give 1,2,3, whereas iloc will only give 1,2 

#### iloc:

```python
# Example: 
points_df= pd.DataFrame(
    {
        '1st Point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd Point':[(-3,4),(0,0), (2,2),(10,10),(1,1)]
    }
)
points_df
```

```python
def euclidean(row):
    pt_A= row['1st Point']
    pt_B= row['2nd Point']

    return round(((pt_A[0]-pt_B[0])**2 + (pt_A[1]- pt_B[1])**2)**0.5,2)

points_df['distance']= points_df.apply(euclidean, axis =1)
points_df
```
