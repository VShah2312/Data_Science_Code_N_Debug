### Bivariate Analysis 

```python
tips = sns.load_dataset('tips')
titanic=df
flights= sns.load_dataset('flights')
iris= sns.load_dataset('iris')
```

```python
# 1. Scatterplot (Num vs Num)
print(tips[['total_bill', 'tip']].corr())
sns.scatterplot(data= tips, x='total_bill', y='tip')
```

```python
# 2. Barplot (Num vs Categorical)

sns.barplot(data=titanic,x ='Pclass', y='Age', hue='Sex')
```

```python
# 3. Crosstab (Cat vs Cat)

pd.crosstab(titanic.Pclass, titanic.Survived)

sns.heatmap(pd.crosstab(titanic.Pclass, titanic.Survived), annot=True)
```

```python
# 4. Pairplot 

sns.pairplot(iris, hue='species')
```

```python
# 5. Lineplot 
flights_df= flights.groupby('year')['passengers'].sum().reset_index()

sns.lineplot(data=flights_df, x='year', y='passengers', marker='o')
```
