### Seaborn: 

```python
import seaborn as sns 
```

```python
# importing inbuilt data set

tips=sns.load_dataset("tips")
iris= sns.load_dataset("iris")
```

```python
sns.scatterplot(data=tips, x='total_bill', y= 'tip') # default axis name is column name
plt.show()
```

```python
# striplot 
"""
Features: 
1. Bivariate Analysis: Categorical vs Numerical 
2. Example: Bill amount vs days 
"""
sns.stripplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
# categorical plot (catplot)
"""
Features: 
1. Bivariate Analysis: Categorical vs Numerical 
2. Mention in kind what kind of chart do you want. 
"""
sns.catplot(data=tips, x='day', y='total_bill', kind= 'strip')
plt.show()
```

```python
# categorical plot (catplot)
"""
Features: 
1. Bivariate Analysis: Categorical vs Numerical 
2. Mention in kind what kind of chart do you want. 
"""
sns.catplot(data=tips, x='day', y='total_bill', kind= 'swarm')
plt.show()

# or 
sns.swarmplot(data=tips, x='day', y='total_bill')
plt.show()
```

```python
tips
```

```python
# Hue: is use to bifercate based on color (hue always has to be categorical)
sns.catplot(data=tips, x='day', y='total_bill', kind= 'swarm', hue='sex')
plt.show()
```

```python
# Boxplot:

""" 
Features: 
1. Its a five point summary.
2. Univariate: Numerical 
3. Bivariate: Categorical vs Numerical. 
"""
# Univariate: 
sns.boxplot(data= tips, x='total_bill')
plt.show()

# Bivariate: 
sns.boxplot(data= tips,x='day', y='total_bill', hue= 'smoker')
plt.show()
```

```python
# Violin plot: 

"""
Features:
1. Combination of boxplot and kde plot.
2. Its a five point summary.
3. Univariate: Numerical 
4. Bivariate: Categorical vs Numerical 
"""
# Univariate: 
sns.violinplot(data= tips, x='total_bill')
plt.show()

# Bivariate: 
sns.violinplot(data= tips,x='day', y='total_bill', hue= 'smoker')
plt.show()
```

```python
# Barplot:

"""
Features:
1. Combination of boxplot and kde plot.
2. Its a five point summary.
3. Bivariate: Categorical vs Numerical (aggregated)
"""
# Bivariate: 
sns.barplot(data= tips, x='sex', y='total_bill', ci= False) # use ci to remove confidence interval
plt.show()

# Bivariate: 
sns.barplot(data= tips,x='day', y='total_bill', hue= 'smoker')
plt.show()
```

```python
# Countplot: 
""" 
Features: 
1. Univariate Analysis: Categorical counts 
2. 
"""
sns.countplot(data=tips, x='sex')
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
# import dataset

tips= px.data.tips()
iris= px.data.iris()
gap= px.data.gapminder()
```

```python
tips.head()
```

```python
iris.head()
```

```python
gap.head()
```

```python
temp_df= gap[gap['year']== 2007]
```

```python
trace_1= go.Scatter(x= temp_df['lifeExp'], y= temp_df['gdpPercap'], mode='markers')
data= [trace_1]
layout= go.Layout(title= 'Life Exp vs GDP per Capita for 2007', xaxis= {'title': 'Life Exp'}, yaxis= {'title':'GDP'})
fig= go.Figure(data, layout)
fig.show()
```

```python
trace_1= go.Scatter(x= temp_df['lifeExp'], y= temp_df['gdpPercap'], mode='markers')
trace_2= go.Scatter(x= [100,150,200], y= [100000, 200000, 500000])
data= [trace_1, trace_2]
layout= go.Layout(title= 'Life Exp vs GDP per Capita for 2007', xaxis= {'title': 'Life Exp'}, yaxis= {'title':'GDP'})
fig= go.Figure(data, layout)
fig.show()
```
