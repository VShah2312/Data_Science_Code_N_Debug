```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
week_day=[1,2,3,4,5,6,7]
temp= [50,51,52,48,47,49,46]
```

```python
plt.plot(week_day,temp)
plt.xlabel('week_day')
plt.ylabel('temp')
plt.title('week vs temp')
plt.show()
```

```python
plt.plot(week_day,temp, color='green', linestyle= '-.', marker= 'o')
plt.xlabel('week_day')
plt.ylabel('temp')
plt.title('week vs temp')
plt.show()
```

```python
days= [1,2,3,4,5,6,7] # weekdays
max_t= [50,51,52,48,47,49,46] # max temp recorder at the day 
min_t= [43,42,40,44,33,35,37] # min temp
avg_t= [45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label= 'max')
plt.plot(days, min_t, label= 'min')
plt.plot(days, avg_t, label= 'avg')
plt.legend() # For legend, you have name all graphs first, we name all graph using label parameter. 

plt.show()
```
