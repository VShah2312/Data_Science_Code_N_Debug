# What is regular expression?
- they basic allow us to search or match a specific text[pattern]

```python
"""
.       - Any Character Except New Line
\d      - Digit (0-9) (lower case gives digits)
\D      - Not a Digit (0-9) (upper case gives non-digits)
\w      - Word Character (a-z, A-Z, 0-9, _) (lower case gives word characters)
\W      - Not a Word Character (upper case gives non-word characters)
\s      - Whitespace (space, tab, newline) (lower case gives whitespace)
\S      - Not Whitespace (space, tab, newline) (upper case gives non-whitespace)

\b      - Word Boundary (lower case gives word boundary)
\B      - Not a Word Boundary (upper case gives non-word boundary)
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)

"""
```

```python
text_to_search = '''
abcdefghijklmnopqurtuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ\s
321-555-4321
1234567890
Ha HaHa
MetaCharacters (Need to be escaped):
. ^ $ * + ? { } [ ] \ | ( )
khanafsaan11.com
321-555-4321
123.555.1234
123*555*-1234
123.555.1234
800-555-1234
900-555-1234
Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T
Mr_hello
'''
```

```python
import re # Regular expression module
```

```python
# raw string to avoid escaping backslashes
print('code\ndebug')

print(r'code\ndebug')  # raw string, no escape sequences processed. 
```

```python
pattern = re.compile(r'a') # put the character/ group of characters that you want to search
matches = pattern.finditer(text_to_search) # text from which it will searched from
# mathces will be an iterable object of match objects
for match in matches:
    print(match)

# or type cast to any iterable object like list
# matches = list(pattern.finditer(text_to_search))
```

```python
# .       - Any Character Except New Line

pattern = re.compile(r'.') 
matches = pattern.finditer(text_to_search) 

for match in matches:
    print(match)
```

```python
# \d      - Digit (0-9)

pattern = re.compile(r'\d') 
matches = pattern.finditer(text_to_search) 

for match in matches:
    print(match)
```

```python
# \D      - not a Digit (0-9)

pattern = re.compile(r'\D') 
matches = pattern.finditer(text_to_search) 

for match in matches:
    print(match)
```

```python
string = 'cat catherine catholic wildcat copycat uncatchable'
```

```python
pattern= re.compile(r'cat') # word boundary before c and after t
matches = pattern.finditer(string)

for match in matches:
    print(match)

# Giving all the words containing cat
```

```python
# \b      - Word Boundary
pattern= re.compile(r'\bcat') # word boundary before c and after t
matches = pattern.finditer(string)

for match in matches:
    print(match)

# Giving 'cat', 'catherine', 'catholic'
```

```python
# \b      - Word Boundary
pattern= re.compile(r'cat\b') # word boundary before c and after t
matches = pattern.finditer(string)

for match in matches:
    print(match)

# Giving 'cat', 'wildcat', 'copycat'
```

```python
# \b      - Word Boundary (one side word character, other side non-word character, sides can be swapped)
pattern= re.compile(r'\bcat\b') # word boundary before c and after t
matches = pattern.finditer(string)

for match in matches:
    print(match)

# Giving only 'cat'. 
```

```python
# \B      - Not Word Boundary (both side has to be word characters)
pattern= re.compile(r'cat\B') # not word boundary before c and after t
# basically it will give the words that have 'cat' in between other characters
# \Bcat will give where the word does not start with 'cat'
# cat\B will give where the word does not end with 'cat'
matches = pattern.finditer(string)

for match in matches:
    print(match)

# Giving only 'uncatchable'. 
```

```python
# search all the 3 digit number and extract it from the text

pattern = re.compile(r'\b\d{3}\b')  # 3 digit number or r'\d\d\d'
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
# search all the phone numbers in the text
# Issue with the above pattern is that it will  match the phone numbers with dashes and dots both as . has other use. 
 
pattern = re.compile(r'\b\d\d\d.\d\d\d.\d\d\d\d\b')
matches = pattern.finditer(text_to_search) 

for match in matches:
    print(match)
```
