Covariance: It is a statistical measure that describes the degree to which to variable are linearly related. It measures how much two variable change together such taht when one variable increases, does the other variable also increases, or does it decrease or no change in the pattern. 

Variance: 
- It measures how much one single variable spreads out around it mean. 
- Example: If you measure the height of people in a group, variance tell you how spread out those heights are. 

Covariance: 
Extends variance to two variables. It measures whether two variables move together or not. 
- Positive covariance: when one increases, the other tends to increase. 
- Negative covariance: when one increases, the other tends to decrease.
- Zero (or near zero): no linear relationship.

 Q: Why is covariance needed? 
 A: 
 1. Variance is a special case of covariance i.e var(x)= cov(x,x) (covariance of variable with iteself is variance of that variable.)
 2. Relationship between variables: Variance alone cannot tell us if two features (say, study hours and exam scores) are related. Covariance reveals whether there is linear dependency. 
 3. This is essential for multivariate stats and ML: 
    - Principal Component Analysis (PCA) (dimensionality reduction)
    - Portfolio Optimization in finance (diversification depends on covariance between assets)
    - Gaussian distributions (multivariate normal is defined by a mean vector + covariance matrix)


Example: 
Imagine tracking height (X) and Weight(Y) in a dataset. 
* Variance of height -> how much people's height vary.
* Variance of weight -> how much people's weight vary. 
* Covariance between height and weight ->  wether taller people also tend to be heavier. 

Variance answers: "How spread out is this one thing?"
Covariance answers: "Do these things move together - and in what direction?"

### Summary of covariance: 
- cov(x,x)= var(x) (variance is special case of covariance)
- Covariance has no fixed bound (-inf, inf)- it can be very large (positive or negative), depending on the data scale. 
- Scaling effect: cov(aX, bY)= ab* cov(X,Y): multiplying variables by constants scales covariance accordingly. Unit of covariance is Unit of X * Unit of Y
- Relative bound (Cauchy- Schwarz inequality): |cov(X,Y)| <= sqrt(var(X) * var(Y)), so covariance magnitude cannot exceed the product of standard deviations. 
- Interpretation: 
    * Large positve: strong upward linear relationship.
    * Large negative: strong downward linear relationship. 
    * Near zero: weak or no linear relationshp.

- Problem: magnitude depends on units (eg: cm vs m, dollars vs cents)
- Solution: use correlation (normalized covariance, always between -1 and 1) for scale free comparison.

# Note: Correlation doesnt imply causation

- The phrase "correlation does not imply causation" means that just beacuase two variables are associated with each other, it does not necessarily mean that one causes the other. In other workds, a correlation between two variables does not necessarily implies that one variable is the reason for the other variables behaviours. 
- Thus, while corelations can provide valuable insights into how different variables are related, they cannot be used to establish causality. Establishing causality often requires additional evidence, such as experiments, randomized controlle trials, or well designed observational studies.
