#### Group by Attributes and Methods 

```python
# How to find number of groups we have: 

len(genres)
# or 

imdb_100['Genre'].nunique()
```

```python
"""
imdb_100.groupby('Genre')
It's a DataFrameGroupBy object, a lazy, iterable object used to apply aggregation functions later.
.shape works on actual DataFrames or Series (which have defined dimensions).
.groupby() returns a GroupBy object, not a DataFrame.
To get the size of each group, you use .size().
"""

"""
imdb_100.groupby('Genre').size()
This returns a Series where:
The index is each unique Genre
The value is the count of rows (i.e. number of movies) in each genre group.
"""
```

```python
# How to find number of rows in each groups: 

genres.size()
# or 
genres['Series_Title'].count()
# or 
```

```python
# To get first row of each group: 
genres.first()
# To get last row of each group 
genres.last()
# Say you want 10th row for each group 
genres.nth(11) # Note: we are missing few genre as we dont have 10th row in them.
```

```python
# get groups:

genres.get_group('Fantasy') # Group by technique

# or
 
imdb_100[imdb_100['Genre']=='Fantasy'] # Filtering technique 

""" 
Which of this method is optimized? 
Optimized is group by technique as you already have access to the groups, thus you iterate it only over groups, where as filtering iterate over whole dataset. 
This group by technique is faster than filtering techniques.
"""
```

```python
# groups : give index of the rows in each groups.
genres.groups
```

```python
# describe: Its same as for dataframe, but gives all information for each groups. 
genres.describe()
```

```python
# sample: same as dataframe, but will give sample for each group. 

genres.sample(3, replace= True) # Here it fails on replace= True, as we have group with less than 3 rows. i.e population < sample requested
# Thus, using replace=True, it will repeat the rows 3 times. Eg: Thriller has only one record, so it will just repeat that row 3 times. 
```

```python
# nunique(): same as dataframe, just give all details for each groups.
genres.nunique()

# note: we dont have unique() in groupby. 
```

```python
# agg method: take in dict, and allow you apply multiple functions to multiple groups. 

# passing dict: 
genres.agg(
    {'Runtime': 'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'min'
    }
)
```

```python
# Allows us to pass functions to aggregate on in one go: 

# passing list: 
genres.agg(['min', 'max', 'sum'])

```

```python
# combined: dict and list
genres.agg(
    {'Runtime': ['min','mean'],
    'IMDB_Rating':'mean',
    'No_of_Votes':['sum','max'],
    'Gross':'sum',
    'Metascore':'min'
    }
)
```

```python
# apply method: same as dataframe, a function you want to apply to each group. 

genres.apply(min) #Each row is the minimum per column for that group.
```

#### Working of joins: 

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})

df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
```

```python
# inner join: 
pd.merge(df1,df2, on='city', how= 'inner')
```

```python
# left join: 
pd.merge(df1,df2, on='city', how= 'left')
```

```python
# right join: 
pd.merge(df1,df2, on='city', how= 'right')
```

```python
# outer join: 
pd.merge(df1,df2, on='city', how= 'outer')
```
